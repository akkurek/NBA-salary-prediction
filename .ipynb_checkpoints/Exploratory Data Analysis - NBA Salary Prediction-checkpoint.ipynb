```python
def slope_coeff(data_x, data_y):
    "This function is written to yield both the slope between two variables and it's Pearson Correlation Coefficient"
    slope, intercept = np.polyfit(data_x, data_y, 1)
    coeff = np.corrcoef(data_x, data_y)[0,1]
    return slope, coeff
```

```python
slope_coeff(df['salary'], df['Point'])
```

For every point scored per game, a player can expect to increase his salary by about $4.500,000, and point has strong correlation with salary.

```python
slope_coeff(df['salary'], df['Assist'])
```

For every assist made per game, a player can expect to increase his salary by about $1.600,000, and assist has strong correlation with salary.

```python
slope_coeff(df['salary'], df['Rebound'])
```

For every rebound collected per game, a player can expect to increase his salary by about $1.450,000, and rebound has strong correlation with salary.

```python
slope_coeff(df['salary'], df['BLK'])
```

```python
new_columns = ['Age', 'Game', 'GS', 'Minutes', 'FG', 'FG%', '3P',
               '3P%', '2P', '2P%', 'eFG%', 'FT', 'FT%', 'ORB', 'DRB',
               'Rebound', 'Assist', 'STL', 'BLK', 'TOV', 'PF', 'Point',
               'salary', 'Years']
df_coeff = df[new_columns]
for column in df_coeff.columns:
    if column != 'salary':
        print('Holding all other features fixed, a 1 unit increase in  {}'.format(column),
              'a player can expect to increase his salary by : {}'.format(int(slope_coeff(df[column], df['salary'])[0])),
              'and the correlation between salary and {}'.format(column),
              'is {}'.format(slope_coeff(df[column], df['salary'])[1]), '\n')
    
```

As a result, through my EDA, I have found that, though there may not be just 1 major independent variable that affects salary, there are many statistics that are interrelated and may affect salary prediciton.

My hypothesis is that Points, Defensive Rebound, Field Goal Made, Minutes Played, Free Throw Made and Minutes per Game are the strongest indicators for how much a player will or should be paid.
