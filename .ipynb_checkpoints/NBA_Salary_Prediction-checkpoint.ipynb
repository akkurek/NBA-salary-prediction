### Removing the '$' signs and commas from the salary columns and converting them to numeric

```python
df18['Salary 2018-19'] = df18['Salary 2018-19'].replace({'\$': '', ',': ''}, regex=True)
df18['Salary 2018-19'] = pd.to_numeric(df18['Salary 2018-19']).astype(int)
```

```python
df19['Salary 2019-20'] = df19['Salary 2019-20'].replace({'\$': '', ',': ''}, regex=True)
df19['Salary 2019-20'] = pd.to_numeric(df19['Salary 2019-20']).astype(int)
```

```python
df19['Salary 2019-20'] = ((df19['Salary 2019-20'] / 1000000) * 1).astype(int)
df18['Salary 2018-19'] = ((df18['Salary 2018-19'] / 1000000) * 1).astype(int)
```

```python
def plotpos(pos):
    data = df19[df19.Pos == pos].sort_values('PTS',ascending=False)
    area = 100 *df19.TRB
    color = df19.MP
    
    df19.plot.scatter('Salary 2019-20','PTS',
                      s=area, c=color,
                      colormap=matplotlib.cm.get_cmap('Purples'), vmin=10, vmax=48,
                      linewidths=1, edgecolors='k', sharex=False)

plotpos('PG')
```

```python
def plotpos(pos):
    dff = df19[df19.Pos == pos].sort_values('PTS',ascending=False)
    area = 100 *dff.TRB
    color = dff.AST
    edgecolor = dff.Pos.map({'C': 'skyblue','PG': 'gold','PF': 'palegreen','SG': 'coral', 'SF' : 'dimgrey'})
    
    dff.plot.scatter('Salary 2019-20','PTS',
                      s=area, c=color,
                      colormap=matplotlib.cm.get_cmap('Purples'), vmin=0, vmax=10,
                      linewidths=1, edgecolors=edgecolor, sharex=False, figsize=(10,6.5) )
plotpos('C')
```

```python
def pearson_r(x, y):
    """Compute Pearson correlation coefficient between two arrays."""
    # Compute correlation matrix: corr_mat
    corr_mat = np.corrcoef(x, y)

    # Return entry [0,1]
    return corr_mat[0,1]

# Compute Pearson correlation coefficient for I. versicolor
r = pearson_r(df18['Salary 2018-19'], df18['PTS'])

# Print the result
print(r)
```
