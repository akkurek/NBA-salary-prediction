## Loading csv files and reading with pandas

```python
# (encoding = "ISO-8859-1") used to avoid encoding errors.
# four files will be uploaded for salary and statistics
stats18 = pd.read_csv("data/player_stats_18.csv", encoding = "ISO-8859-1" )
stats19 = pd.read_csv("data/player_stats_19.csv", encoding = "ISO-8859-1" )
salary18 = pd.read_csv("data/player_salary_18.csv", encoding= "ISO-8859-1")
salary19 = pd.read_csv("data/player_salary_19.csv", encoding= "ISO-8859-1")
```

```python
#These are the players called rookie, who will be dropped from the actual list if they have 3 or less years experience
#because these players have 3=years-rookie contracts even if they are having high points for each game, 
#their salaries won't be reflecting reality.
files = ['data/drafted_players_19.csv', 'data/drafted_players_18.csv', 'data/drafted_players_17.csv', 'data/drafted_players_16.csv']
df =[]
for file in files:
    data = pd.read_csv(file, encoding = "ISO-8859-1", header=1)
    df.append(data)
drafted = pd.concat(df)
drafted = drafted[['Player', 'Yrs']]
drafted.dropna(how='any', inplace=True)
drafted.Yrs.count() #number of players that still have rookie contracts which we need to drop from our 2018-19 lists.
```

### Removing the '$' signs and commas from the salary columns and converting them to numeric

```python
df18['Salary 2018-19'] = df18['Salary 2018-19'].replace({'\$': '', ',': ''}, regex=True)
df18['Salary 2018-19'] = pd.to_numeric(df18['Salary 2018-19']).astype(int)
```

```python
df19['Salary 2019-20'] = df19['Salary 2019-20'].replace({'\$': '', ',': ''}, regex=True)
df19['Salary 2019-20'] = pd.to_numeric(df19['Salary 2019-20']).astype(int)
```

```python
df19['Salary 2019-20'] = ((df19['Salary 2019-20'] / 1000000) * 1).astype(int)
df18['Salary 2018-19'] = ((df18['Salary 2018-19'] / 1000000) * 1).astype(int)
```
